# Secuencias
En Python las secuencias son las colecciones ordenadas de distintos tipos de datos, ya sean similares o diferentes. En las secuencias podemos almacenar m√∫ltiples valores de forma organizada y eficiente. 

Existen distintos tipos de secuencias: 

* String: 
Cadenas que representan uno o varios caracteres. B√°sicamente se trata de una l√≠nea de texto, que se encuadra dentro de la etiqueta str

* Listas:
 Las listas son, fundamentalmente, lo que conocemos como arrays en otros lenguajes de programaci√≥n. Es un elemento flexible ya que no es necesario que todos sus componentes sean del mismo tipo. 

* Tupla: Igual que sucede con las listas, la tupla tambi√©n es una colecci√≥n ordenada de objetos en Python. Se diferencian en que las tuplas son inmutables, mientras que las lista se pueden modificar.

# Set
El tipo de dato Set (conjuntos) en Python se emplea para agrupar los objetos √∫nicos y que no est√°n ordenados. 

En los conjuntos en Python, los valores duplicados se eliminan y solo se conservan los valores √∫nicos. Operaciones como intersecciones y uniones se pueden realizar entre uno o m√°s conjuntos. 

# Diccionario:

# Entrada del usuario
Puede codificarlo/textearlo para que el programa le diga al usuario que ingrese informaci√≥n. 

Guarde los datos en el programa y act√∫e en consecuencia.

Para capturar informaci√≥n del usuario, utilice la sentencia input. Aqu√≠ hay un ejemplo: input(), veamos el error clasico de la calculadora mas malota ü§ù

```python
# Funci√≥n input para procesar la enterda del usuario 
numero_1 = input('Escribe el primer: ')
numero_2 = input('Escribe el segundo n√∫mero: ')

resultado = numero_1 + numero_2

print('El resultado de los dos n√∫meros es: ', resultado)
type(numero_1)
```
